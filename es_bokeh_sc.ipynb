```python
u, r, RA, Dec = hdul[1].data['umag'], hdul[1].data['rmag'], hdul[1].data['RAJ2000'], hdul[1].data['DEJ2000']

ur= u- r
```

```python
from bokeh.io import output_notebook, show 
from bokeh.layouts import gridplot
from bokeh.plotting import figure

# Need this line to display bokeh plots inline in the notebook
output_notebook()
import numpy as np
from bokeh.layouts import column
from bokeh.models import Slider
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, Range1d
```

```python
source = ColumnDataSource(data=dict(u=u,r=r,RA=RA,Dec=Dec,ur=ur))
TOOLS = "box_zoom,box_select,lasso_select,reset"

kwargs = {'selection_fill_color':'steelblue', 
          'selection_line_color': 'steelblue',
          'nonselection_fill_color': 'silver', 
          'nonselection_line_color': 'silver'}


left = figure(tools=TOOLS, plot_width=300, plot_height=300,title='spatial distribution')

left.circle('RA', 'Dec', source=source, **kwargs)
left.x_range = Range1d(100.6, 99.8)
left.xaxis.axis_label = 'Ra'
left.yaxis.axis_label = 'Dec'

right = figure(tools=TOOLS, plot_width=300, plot_height=300, title='Color-Magnitude diagram')
right.square('ur', 'u', source=source, **kwargs)
right.xaxis.axis_label = 'u -r'
right.yaxis.axis_label = 'u'

p = gridplot([[left, right]])

show(p)
```
